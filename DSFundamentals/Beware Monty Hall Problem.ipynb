The Monty Hall problem is an example of conditional probability and paints a good picture of how intuitions of statistics can be challenged. 

To start with this problem did not make much sense to me, but after reading through the wiki article I've managed to gather a pretty good grasp on what it's all about. 

Here's a summary of my attempt at articulating the reasons why switching the initial guess gives a higher probability of choosing the car. 

Initially, the player has a 1/3 chance of choosing the car. 
After the initial guess, the conditions for revealing one of the non-car choices alters the probabilities of what lies behind each door. 

1. If the car is initially selected, hall can choose between either of the other doors to open
2. If a non-car is selected, hall is forced to show the only remaining non-car. 

SO, the initial probability of selecting the car is 1/3. That means there is a 2/3 change the car is behind the other two doors. After the reveal, probability of that you're choice is still 1/3. However, because the revealed door was 1/2 of the possible 2/3 of "car" outcomes, if you switch your choice you how have a probability of 2/3 in selecting the car. 

This most important part of this problem is understanding conditional probability and the use of new information in the altering of probability.
