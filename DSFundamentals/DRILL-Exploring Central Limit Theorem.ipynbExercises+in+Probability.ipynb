1. Calculate the probability of flipping a balanced coin four times and getting each pattern: HTTH, HHHH and TTHH.

```python
# for all these patterns, the possible outcomes are independent. This means the probability of each pattern
# is the product of each event. 
prob_for_all = 1/2 ** 4
print(prob_for_all)
```

2\. If a list of people has 24 women and 21 men, then the probability of choosing a man from the list is 21/45. What is the probability of not choosing a man?


```python
# The probability of NOT choosing a man is simply the number of possible event that are not men. 
total = 45
men = 21
not_men = total - men 
prob_not_men = not_men / total
print("24/45 or {}".format(prob_not_men))
```
