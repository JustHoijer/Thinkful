Drill: Present the same data multiple ways

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


# Loading an example dataset provided by the seaborn package.
flights = sns.load_dataset("flights")
print(flights.head(n=4))
```

### Simple Histogram

```python
JanFlights = flights.loc[(flights['month'] == 'January'), 'passengers']
FebFlights = flights.loc[(flights['month'] == 'February'), 'passengers']
MarFlights = flights.loc[(flights['month'] == 'March'), 'passengers']
JanFlights.head()
```

```python
plt.hist(JanFlights, alpha=.7, label = "Jan Flights")
plt.hist(FebFlights, alpha=.7, label = "Feb Flights")
plt.hist(MarFlights, alpha=.7, label = "Mar Flights")
# plt.xlabel('Passengers per flight')
plt.legend(loc='upper right')
plt.title('Plot 1: # of passengers per month')
plt.show()
```

### Scatter Plots

```python
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
g = sns.lmplot(y = 'horsepower',
               x = 'mpg', 
               data = mpg,
               fit_reg = True,
               scatter_kws={'alpha':0.3})
g.set_ylabels('horsepower')
g.set_xlabels('mpg')

               
plt.title('Scatterplot: mpg vs horsepower')
plt.show()
```

### Subplots

```python
sns.set(style='dark')

g2 = sns.lmplot(y = 'mpg',
               x = 'horsepower',
               hue = 'origin',
               data = mpg,
               col = 'cylinders',
               fit_reg = True,
               ci = False, 
               #col_wrap = 3,
               #height=3, 
               scatter_kws={'alpha':.6},
               aspect=.5)
g2.set_ylabels = 'mpg'
g2.set_xlabels = 'horsepower'
plt.subplots_adjust(left=.125, right=0.9, top=0.9, bottom=0.2, wspace = 0.2, hspace = 0.2)
plt.show()
```

### Bar Plot

```python
sns.set_style('white', {"axes.facecolor": ".9"})

g3 = sns.factorplot(x= 'origin', y='mpg', data=mpg, 
                   size = 7, kind='bar', ci=95)
g3.despine(left=True)
g3.set_ylabels("mpg")
g3.set_xlabels("")
plt.title("Plot: mpg by origin")
plt.show()
```

### Point Plot

```python
sns.set_style('dark', {'axes.edgecolor': '.8'})

g4 = sns.factorplot(x ='mpg', y = 'origin', hue='model_year', data=mpg, 
                   size = 10, kind='point', palette='pastel', ci=95, dodge=True, join=False)
g4.despine(left=True)
g4.set_ylabels("origin")
g4.set_xlabels("mpg")
plt.title("mpg by model year and origin")
plt.show()
```

### Box Plot


```python
g5 = sns.boxplot(x='origin',y='mpg',hue='model_year',data=mpg,palette='pastel')  
sns.set(rc={'figure.figsize':(30,15)})
plt.title('mpg by origin and model_year')

sns.despine(offset=10, trim=False)
g5.set(xlabel='', ylabel='mpg')
plt.legend(loc='best')
plt.show()
```

### Violin plot

```python
sns.set_style('ticks')
sns.set(rc={'figure.figsize':(10,15)})
sns.violinplot(data=mpg, x='origin', y='mpg', hue='origin', inner='points', aspect=0.3)
sns.despine()
#plt.legend(loc='best')
#lt.show()
```
