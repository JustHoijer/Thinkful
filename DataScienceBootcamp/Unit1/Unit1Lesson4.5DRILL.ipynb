Look back to the potential experiments in the previous assignment. For one of those experiments, write up the essential points of a research proposal for an improved version of that experiment.


Experiment:
The Sith Lords are concerned that their recruiting slogan, "Give In to Your Anger," isn't very effective. Darth Vader develops an alternative slogan, "Together We Can Rule the Galaxy." They compare the slogans on two groups of 50 captured droids each. In one group, Emperor Palpatine delivers the "Anger" slogan. In the other, Darth Vader presents the "Together" slogan. 20 droids convert to the Dark Side after hearing Palpatine's slogan, while only 5 droids convert after hearing Vader's. The Sith's data scientist concludes that "Anger" is a more effective slogan and should continue to be used.

### Research Proposal:

- The problem
- The potential solution
- The method of testing

Sith Lords use a slogan to recruit droids into their army, and without an effective slogan, the Dark Side will be unable to successfully rule the galaxy. To ensure the effectiveness of the current slogan, a new one slogan is tested and compared as a possible replacement. Both Sith Lords will deliver each of the slogans to different groups of captured droids. Each group of droids, or sample, will be a randomly selected collection from the same source (or captured group) and include the same number of droids. At the end of each session the number of droids recruited, which slogan was used, and which Sith Lord delivered it will be recorded. These metrics will be used together to determine the effectiveness of each slogan by looking at the average recruitment for each Sith Lord. The interpretation and analysis of these results will determine which slogan should be used.
