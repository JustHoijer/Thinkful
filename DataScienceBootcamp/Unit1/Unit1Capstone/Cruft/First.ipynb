```python
df1 = pd.read_csv("./dutch-energy/Electricity/liander_electricity_01012009.csv")
df1['year'] = '2009'
df2 = pd.read_csv("./dutch-energy/Electricity/liander_electricity_01012010.csv")
df2['year'] = '2010'
df3 = pd.read_csv("./dutch-energy/Electricity/liander_electricity_01012017.csv")
df3['year'] = '2017'
df4 = pd.read_csv("./dutch-energy/Electricity/liander_electricity_01012018.csv")
df4['year'] = '2018'
df = pd.concat([df1,df2,df3,df4])


Nine = df.loc[(df['city']=='AMSTERDAM')&(df['year']=='2009'),'annual_consume']
Ten = df.loc[(df['city']=='AMSTERDAM')&(df['year']=='2010'),'annual_consume']
Seventeen = df.loc[(df['city']=='AMSTERDAM')&(df['year']=='2017'), 'annual_consume']
Eighteen = df.loc[(df['city']=='AMSTERDAM')&(df['year']=='2018'), 'annual_consume']

Nine = Nine.reset_index(drop=True)
Ten = Ten.reset_index(drop=True)
Seventeen = Seventeen.reset_index(drop=True)
Eighteen = Eighteen.reset_index(drop=True)
```

```python
First = pd.DataFrame()
for n in range(10000):
    i = random.randint(1, 15000)
    diff = pd.DataFrame({'delta':Ten[i] - Nine[i]}, index=[i])
    First = pd.concat([First, diff])    
Second = pd.DataFrame()
for n in range(10000):
    i = random.randint(0, 15000)
    diff = pd.DataFrame({'delta':Eighteen[i] - Seventeen[i]}, index=[i])
    Second = pd.concat([Second, diff])
Third = pd.DataFrame()
for n in range(10000):
    i = random.randint(0, 15000)
    diff = pd.DataFrame({'delta':Eighteen[i] - Nine[i]}, index=[i])
    Third = pd.concat([Third, diff])    
```

```python
First = First.reset_index()
First['period'] = 1
Second = Second.reset_index()
Second['period'] = 2
Third = Third.reset_index()
Third['period'] = 3
df = pd.concat([First,Second,Third])
```

```python
df.head()
```

```python
Firstperiod = df.loc[(df['period']==1), 'delta']
Secondperiod = df.loc[(df['period']==2), 'delta']
Thirdperiod = df.loc[(df['period']==3), 'delta']
```

```python
plt.hist(Firstperiod, color='red', alpha=0.4, label='firstyeardifference')
plt.hist(Secondperiod, color='blue', alpha=0.6, label='lastyeardifference')
plt.hist(Thirdperiod, color='black', alpha=0.5, label='fulldifference')
plt.xlabel = 'annual_consumption'
plt.legend(loc='upper right')
plt.title('Plot 1')
plt.show()
```

```python
sns.set(style="whitegrid")

ax = sns.boxplot(x='period',y='delta',data=df,palette='pastel')  

plt.title('Plot 2')

sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Consumption')

plt.show()
```

```python
sns.set(style="darkgrid")

g = sns.factorplot(x="period", y="delta", data=df,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("ConsumptionDifferences")
g.set_xlabels("")
plt.title('Barplot')
plt.show()
```

```python
sns.set(style="whitegrid")

g = sns.factorplot(x="period", y="delta", data=df,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("ConsumptionDifference")
g.set_xlabels("")
plt.title('Pointplot')
plt.show()
```

```python
sns.set(style="white")
g = sns.lmplot(y='delta', # Variable 1.
               x='index', # Variable 2.
               data=First, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4},
              x_jitter=10E5,
              y_jitter=10E2) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("ConsumptionDifference")
g.set_xlabels("Time")
plt.title('Scatterplot')
plt.subplots_adjust(top=0.9)
plt.show()

g = sns.lmplot(y='delta', # Variable 1.
               x='index', # Variable 2.
               data=Second, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4},
              x_jitter=10E5,
              y_jitter=10E2) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("ConsumptionDifference")
g.set_xlabels("Period")
plt.title('Scatterplot')
plt.subplots_adjust(top=0.9)
plt.show()
g = sns.lmplot(y='delta', # Variable 1.
               x='index', # Variable 2.
               data=Third, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4},
              x_jitter=10E5,
              y_jitter=10E2) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("ConsumptionDifference")
g.set_xlabels("Period")
plt.title('Scatterplot')
plt.subplots_adjust(top=0.9)
plt.show()
```
