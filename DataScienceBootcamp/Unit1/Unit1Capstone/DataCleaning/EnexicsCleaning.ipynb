```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
df1 = pd.read_csv("../dutch-energy/Electricity/enexis_electricity_01012010.csv")
df1['year'] = '2010'
df2 = pd.read_csv("../dutch-energy/Electricity/enexis_electricity_01012012.csv")
df2['year'] = '2012'
df3 = pd.read_csv("../dutch-energy/Electricity/enexis_electricity_01012015.csv")
df3['year'] = '2015'
df4 = pd.read_csv("../dutch-energy/Electricity/enexis_electricity_01012018.csv")
df4['year'] = '2018'
df = pd.concat([df1,df2,df3,df4])
```

```python
Ten = df.loc[(df['city']=='GRONINGEN')&(df['year']=='2010'),'annual_consume']
Twelve = df.loc[(df['city']=='GRONINGEN')&(df['year']=='2012'),'annual_consume']
Fifteen = df.loc[(df['city']=='GRONINGEN')&(df['year']=='2015'), 'annual_consume']
Eighteen = df.loc[(df['city']=='GRONINGEN')&(df['year']=='2018'), 'annual_consume']

Ten = Ten.reset_index(drop=True)
Twelve = Twelve.reset_index(drop=True)
Fifteen = Fifteen.reset_index(drop=True)
Eighteen = Eighteen.reset_index(drop=True)
```
