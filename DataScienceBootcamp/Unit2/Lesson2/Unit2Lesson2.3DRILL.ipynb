## Unit 2 Lesson 2.3 Drill: Regression or classification 


1\. The amount a person will spend on a given site in the next 24 months.

This example requries a regression model. The variables involved are continous (money spent, and time). Both of these variables have a defined zero and related intervals. 


2\. What color car someone is going to buy.

A classification model would be used here. Color is a categorical variable, and the outcome could be represented as either binary or probabilisitically. 


3\. How many children a family will have.

Regression model, the classifier in this case is a continous variable. 


4\. If someone will sign up for a service.

Classification model: classifier is either a yes or no. 


5\. The number of times someone will get sick in a year.

Regression model, includes continous variables with set intervals. 


6\. The probability someone will get sick in the next month.

Classifier model would work, using the binary input variables and a probability outcome. |  

7\. Which medicine will work best for a given patient.

This one could be either regression or classificiation. I would guess that a classification model would be best though. It would be a simpler version to assume the outcome of "best" would be based on a binary variable. 

The regression model could be used in determining a range of "best".
