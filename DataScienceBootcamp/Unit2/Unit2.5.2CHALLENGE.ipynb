```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn import linear_model
import seaborn as sns
%matplotlib inline
```

```python
raw_14 = pd.read_excel('Table_8_Offenses_Known_to_Law_Enforcement_by_New_York_by_City_2014.xls',
                         skiprows=4)
features1 = raw_14[['Population', 'Property\ncrime','Violent\ncrime', 'Robbery', 'Burglary', 'Murder and\nnonnegligent\nmanslaughter' ]]
features1['Murders'] = features1['Murder and\nnonnegligent\nmanslaughter']
features1['PopSquared'] = features1['Population']**2


raw_13 = pd.read_excel('table_8_offenses_known_to_law_enforcement_new_york_by_city_2013.xls', skiprows=4)
features2 = raw_13[['Population', 'Property\ncrime','Violent\ncrime', 'Robbery', 'Burglary', 'Murder and\nnonnegligent\nmanslaughter' ]]
features2['PopSquared'] = features2['Population']**2
features2 = features2.dropna(axis=0)
features2['Murders'] = features2['Murder and\nnonnegligent\nmanslaughter']


```

```python
series = features2['Murders']
x = 0
for i in series:
    
    if i == 0:
        features2['Murders'][x] = 0
        x+=1
    else:
        features2['Murders'][x] = 1
        x+=1
series = features2['Robbery']
x = 0
for i in series:
    
    if i == 0:
        features2['Robbery'][x] = 0
        x+=1
    else:
        features2['Robbery'][x] = 1
        x+=1
series = features2['Burglary']
x = 0
for i in series:
    
    if i == 0:
        features2['Burglary'][x] = 0
        x+=1
    else:
        features2['Burglary'][x] = 1
        x+=1
```

```python
series = features1['Murders']
x = 0
for i in series:
    
    if i == 0:
        features1['Murders'][x] = 0
        x+=1
    else:
        features1['Murders'][x] = 1
        x+=1
series = features1['Robbery']
x = 0
for i in series:
    
    if i == 0:
        features1['Robbery'][x] = 0
        x+=1
    else:
        features1['Robbery'][x] = 1
        x+=1
series = features1['Burglary']
x = 0
for i in series:
    
    if i == 0:
        features1['Burglary'][x] = 0
        x+=1
    else:
        features1['Burglary'][x] = 1
        x+=1
```

```python
data = features1
outlier = data.Population.max()
data.loc[data.Population == outlier]
data = data[data['Population']< 8000000]
data.dropna()
```

```python
data.describe()
```

```python
regr = linear_model.LinearRegression()
Y = data['Violent\ncrime'].values.reshape(-1, 1)
X = data[['Population', 'PopSquared','Murders', 'Robbery', 'Burglary', ]]
regr.fit(X, Y)

# Inspect the results.
print('\nCoefficients: \n', regr.coef_)
print('\nIntercept: \n', regr.intercept_)
print('\nR-squared:')
print(regr.score(X, Y))
```
