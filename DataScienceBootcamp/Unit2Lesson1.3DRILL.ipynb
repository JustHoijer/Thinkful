1\. What is the 1994 rate of juvenile delinquency in the U.S.?

Altough alittle simple, this seems like a "good" research question. Provided the data exists, a simple mathematical model can be used to determine the rate of juvenile dlinquency. Using a 'per time' or 'per standard' approach, this question can be answered using a numeric solution. 

2\. What can we do to reduce juvenile delinquency in the U.S.?

This question seems alittle vague. There might be too many factors that contribute to juvenile delinquency for a mathematical model to accuracly answer this question. Given data related to attributing factors, one might be able to determine which facor/factors can be focused on. 


3\. Does education play a role in reducing juvenile delinquents' return to crime?

This is definitly a "good" research question. It is both direct and mathemically solvable. Given the data, a relationship can be easily drawn between education proceeding crime and a return to crime. The results could include probabilities or quantities. 


4\. How many customers does AT&T currently serve in Washington, DC?

This is a "bad" research question. Just a number from a dataset doesn't require any mathematical modeling. Even if the data population was larger than just the D.C area, the answer is not one that requires any statistic modeling. An alternative might be to determine the proportion of people AT&T serves in the area.

Again, given that data exists for the current customer acquisition techiques; This seems like a "good" research question. By determining how resources can be better allocated, a proper action plan (including probabilities) can be created for the attraction of new customers.
