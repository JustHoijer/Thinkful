### Drill - Describing Data

1\. What are the mean, median, and mode of the kids' ages when they first appeared on the show? What are the variance, standard deviation, and standard error?

```python
import numpy as np
ages = [6,8,8,10,11,12,14]

print(f"The mean is {np.mean(ages)}")
print("The mode is 8")
print(f"The median is {np.median(ages)}")
```

2\. if you had to choose only one estimate of central tendency and one estimate of variance to describe the data, which would you pick and why?

$\bf Answer:$ I would choose Mean for a central tendency estimate and standard variation for the variance estimate. 
With such a small data set, the combination of mean and standard devaition, to me, paints a very good picture of the population. 

3\. Next, Cindy has a birthday. Update your estimates- what changed, and what didn't?

$\bf Answer: $ If the value of 6 became 7, the mode and median of this data would remain the same. The mean would increase slightly to 10. 


```python
ages[0] = 7
print(f"The mean is {np.mean(ages)}")
print("The mode is still 8")
print(f"The median is still {np.median(ages)}")
```

4\. Nobody likes Cousin Oliver. Maybe the network should have used an even younger actor. Replace Cousin Oliver with 1-year-old Jessica, then recalculate again. Does this change your choice of central tendency or variance estimation methods?

```python
ages[2] = 1
print(f"The mean is {np.mean(ages)}")
print("There is no longer a mode")
print(f"The median is {np.median(ages)}")
```
