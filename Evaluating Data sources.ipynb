Drill
In each of the scenarios, find possible shortcomings of the theoretical or actual data sources to answer the given question. What could be done to either adjust the analysis or reframe the question so that you can answer it accurately?

- Data Source: Amsterdam availability data scraped from AirBnB on December 24th. Question: What are the popular neighborhoods in Amsterdam?

This data was collected on Christmas Eve, which is a exceptional circumstance that would constitute a change of question. In this case, I would change the question to: Which neighborhoods are most appealing to christmas vacationers?

- Data Source: Mental health services use on September 12, 2001 in San Francisco, CA and New York City, NY. Question: How do patterns of mental health service use vary between cities?

Again, this data was collected under expetional circumstances. The day after the 9/11 attacks would suggestivly alter the use of mental health services. 
Possible changes of question: 

How did the 9/11 attacks affect the mental health of two large American cities?

How did proximity to the attacks affect the increase in use of mental health services?

- Data Source: Armenian Pub Survey. Question: What are the most common reasons Armenians visit local pubs?

In this case, I'd sugegst there would be a lack of quality in data or useful questions. There isn't any specific direction that the data/ question points in. Perhaps, also, there is a bias to the collection method. This data wouldn't include people who don't like to visit pubs?
