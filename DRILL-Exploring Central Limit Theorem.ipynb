```python
import numpy as np
import matplotlib.pyplot as plt
print("Population of 100:\n ")
n1, p1, size1 = 10, 0.2, 100
pop1 = np.random.binomial(n1, p1, size1)
mean1 = pop1.mean()
std1 = pop1.std()
print(f"Mean of pop1 :{mean1} \nStandard Variation of pop1: {std1}\n")
n2, p2, size2 = 10, 0.5, 100
pop2 = np.random.binomial(n2, p2, size2)
mean2 = pop2.mean()
std2 = pop2.std()
print(f"Mean of pop2 :{mean2} \nStandard Variation of pop2: {std2}")
```

```python
print("Population of 10000: \n")
n1, p1, size1 = 10, 0.2, 10000
pop1 = np.random.binomial(n1, p1, size1)
mean1 = pop1.mean()
std1 = pop1.std()
print(f"Mean of pop1 :{mean1} \nStandard Variation of pop1: {std1}\n")
n2, p2, size2 = 10, 0.5, 10000
pop2 = np.random.binomial(n2, p2, size2)
mean2 = pop2.mean()
std2 = pop2.std()
print(f"Mean of pop2 :{mean2} \nStandard Variation of pop2: {std2}")
```

## Changing population size, effect on mean and std

By changing the population size from 100 to 10,000, there are minor changes to the mean and standard deviation. Noteably, an increase in population led to a slight decrease in both the standard deviation and mean. 

```python
print("pop 1 with p-value of 0.3: \n")
n1, p1, size1 = 10, 0.3, 10000
pop1 = np.random.binomial(n1, p1, size1)
mean1 = pop1.mean()
std1 = pop1.std()
print(f"Mean of pop1 :{mean1} \nStandard Variation of pop1: {std1} \n" )
n2, p2, size2 = 10, 0.5, 10000
pop2 = np.random.binomial(n2, p2, size2)
mean2 = pop2.mean()
std2 = pop2.std()
print(f"Mean of pop2 :{mean2} \nStandard Variation of pop2: {std2}")
```

## Changing of p-value, effect on descriptive statistics 

In both populations, the corresponding mean is directly proportional to the p-value. An increase or decrease will directly to a porpotional change in mean. 
Interestly, the standard deviation also increased.
