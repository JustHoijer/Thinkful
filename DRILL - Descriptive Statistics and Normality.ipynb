```python
import numpy as np
import matplotlib.pyplot as plt
distribution = np.random.normal(0, 1, 100)
```

```python
mean = distribution.mean()
std = distribution.std()
plt.hist(distribution)
plt.axvline(distribution.mean(), color='r')
plt.axvline(distribution.std(), color='r')
plt.show()
print(f"Mean is :{mean} \nStandard Deviation: {std}")
```

When looking at this plot, it's clear to me that the value of standard deviation serves little value on this graph. Compared to the obvious nature of the mean (occuring near the middle of the distribution), the std's location on the graph is neither visually neccesary or made easier to understand. This is not to say the value of standard deviation is not important, but only to suggest it's purpose on the graph seems unnessary. 

```python
# first distribution
mean2, std2 = 5, 0.5
dis = np.random.normal(mean2, std2, 1000)

# second distribution 
mean3, std3 = 10, 1
dis2 = np.random.normal(mean3, std3, 1000)

# combination 
dis3 = dis + dis2
plt.hist(dis3)
plt.axvline(dis3.mean(), color='r')
plt.axvline(dis3.std(), color='r')
plt.show()
```
